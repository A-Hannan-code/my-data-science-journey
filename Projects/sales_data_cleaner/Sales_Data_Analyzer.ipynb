```python
import pandas as pd
import numpy as np

# Loading Data......
df = pd.read_csv('sales_data_raw.csv')

# Analysing Data..........
print("------ Data -----\n")
print(df.head(12))

# Missing Values......
print("\n------- Null --------")
print(df.isnull().sum())

# INFO.........
print("\n------- INFO --------")
print(df.info())

# ...........................................................
# Filtering Date.........

df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df["Date"] = df["Date"].bfill()

# Filtering Customers.........
df["Customer"] = df["Customer"].str.strip()
df["Customer"] = df["Customer"].fillna("Unknown")
df["Customer"] = df["Customer"].str.capitalize()

# Filtering Products......
df["Product"] = df["Product"].fillna("Unknown Product")


# Filtering Category......... 
df["Category"] = df["Category"].fillna("Unknown Category")
df["Category"] = df["Category"].str.lower()
# print(df["Category"].unique())
# print(df["Category"].value_counts())

df["Category"] = df["Category"].replace({
    "clothng": "clothing",
    "stationary": "stationery"
})
df["Category"] = df["Category"].str.capitalize()

# Filtering Quantity........
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].mean()).round(0)
df["Quantity"] = df["Quantity"].ffill().abs()


# Filtering Price......
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
df["Price"] = df["Price"].fillna(df["Price"].mean()).round(0).abs()


# Adding Column.........
df["Revenue"] = df["Quantity"] * df["Price"]

# Checking Data After Cleaning....
df.head(12)
df.info()
df.describe()
# Saving Data to a File.....

df.to_csv("sales_data_clean.csv", index=False)
print("✅ Cleaned data saved as sales_data_clean.csv")
```
