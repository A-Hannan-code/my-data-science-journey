## Data Transformation

```python
import pandas as pd

df = pd.read_csv('students.csv')
df
```

```python
df.sort_values('Age')
```

```python
df.sort_values('Math', ascending=False)
```

```python
df2 = df.sort_values(['Science', 'Age'], ascending=False).copy()
df2
```

```python
df2.reset_index()
```

```python
df2.reset_index(drop=True, inplace=True)
```

```python
df2
```

```python
df2['Rank'] = df2['Math'].rank(ascending=False)
```

```python
df2
```

```python
df2.rename(columns={"Rank": 'Ranking'}, inplace=True)
```

```python
df2
```

```python
df2.rename(index={0: 'row1'}, inplace=False)
```

```python
df2.rename(index={'row1': 0}, inplace=True)
```
